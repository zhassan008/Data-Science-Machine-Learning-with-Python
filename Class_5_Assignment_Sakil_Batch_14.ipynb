```python
import pandas as pd
df = pd.read_csv('supershops.csv')  # Read Data Table
```

Standard Scaler using LOOP

```python
from sklearn.preprocessing import StandardScaler
std = StandardScaler()  
df_std = df.copy()    #Copy Data Table for Standard Scaler Method
```

```python
x_std = df_std.drop(columns = 'Profit')  #Separate x from Data Table
x_std.head()
```

```python
y_std = df_std[['Profit']]   #Separate y from Data Table
y_std.head()
```

```python
for column in x_std:  # Loop to Transform all Column
    if x_std[column].dtype == 'float64'or x_std[column].dtype == 'int64':
        x_std[column] = std.fit_transform(x_std[[column]])  # Fit and Transform by Standard Scaler(std) method
x_std.head()
```

Min Max Scaler

```python
x_mms['Administration'] = mms.fit_transform(x_mms[['Administration']]) #Fit and Transform 'Administration' Column by Min Max Scaler
x_mms.head()
```

```python
x_mms['Transport'] = mms.fit_transform(x_mms[['Transport']]) #Fit and Transform 'Administration' Column by Min Max Scaler
x_mms.head()
```
